# Исследование продаж видеоигр

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**В ходе работы нам надо будет составить портрет пользователя каждого региона и определить:**
1. _Самые популярные платформы (топ-5). Описать различия в долях продаж._
2. _Самые популярные жанры (топ-5). Пояснить разницу._
3. _Влияет ли рейтинг ESRB на продажи в отдельном регионе?_

**Проверить гипотезы:**
1. _Средние пользовательские рейтинги платформ Xbox One и PC одинаковые_
2. _Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные._

## Подготовка данных

Приводим названия столбцов к нижнему регистру:

````python
df.columns = df.columns.str.lower()
df.head(2)
````

В датафрейме есть 2 строки с пропущенными значениями в "name":

Возможности восстановить их нет. Можно смело удалить. К тому же уйдут пропущенные значения из столбца "genre".

````python
df = df.dropna(subset=['name'])
````

Восстановить пропуски в столбце "year_of_release" есть возможность. Например, дата выхода спортивных симуляторов не соответсвует году, который есть в их названиях ("Fifa 2003" вышла в 2002 г.) Однако, используя данную закономерность, заполнится лишь их малая часть. Простым и эффективным решением будет удалить строки с пропущенными значениями.

````python
df = df.dropna(subset=['year_of_release'])
````

````python
df['year_of_release'] = pd.to_datetime(df['year_of_release'], format = '%Y')
df['year_of_release'] = pd.DatetimeIndex(df['year_of_release']).year
````

````python
df['critic_score'] = (
    df['critic_score']
    .fillna(-1) #заменим на -1
    .astype('float64')
)
````

Аббревиатура "tbd" переводится как "требует уточнения". Потенциально, игры с таким значением в столбце "оценка пользователей", оценены, но они не дошли до датасета. Такие значения лучше заменить не "0", а "-1". Это своеобразная "метка". Пропуски лучше заменить на "0". 

````python
df['user_score'] = (
    df['user_score']
    .fillna(0)
    .replace('tbd', -1)
    .astype('float')
)
````

В столбце "рейтинг" пропущенные значения стоит заменить на "неизвестно".

````python
df['rating'] = (
    df['rating']
    .fillna('unknown')
    .astype('str')
)
````

Часть пропущенных значений в столбцах "оценка критиков" и "оценка пользователей" можно объяснить тем, что в XX веке доступность большей части пользователей к интернету была невозможна. Оценки просто-напросто было некому. Скорее всего, и критикам оценки было ставить негде. 

Создадим новый столбец, в котором отразим общие продажи.

````python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales'] 
````

### Диаграмма рассеяния и корреляция между отзывами и продажами. 

Так как пропуски в столбцах отзывов пользователей и критиков были заменены на "0" и "-1". Необходимо будет отфильтровать датасет:

````python
actual_per_user_df = (
    actual_per_df
    .query('user_score > 0')
)

actual_per_critic_df = (
    actual_per_df
    .query('critic_score > 0')
)
````

````python
corr_user3DS = actual_per_user_df[actual_per_user_df['platform'] == '3DS']
corr_critic3DS = actual_per_critic_df[actual_per_critic_df['platform'] == '3DS']
print('Корреляция между отзывами пользователей и продажами 3DS:', corr_user3DS['user_score'].corr(corr_user3DS['total_sales']))
print('Корреляция между отзывами критиков и продажами 3DS:', corr_critic3DS['critic_score'].corr(corr_critic3DS['total_sales']))

corr_userXOne = actual_per_user_df[actual_per_user_df['platform'] == 'XOne']
corr_criticXOne = actual_per_critic_df[actual_per_critic_df['platform'] == 'XOne']
print('Корреляция между отзывами пользователей и продажами XOne:', corr_userXOne['user_score'].corr(corr_userXOne['total_sales']))
print('Корреляция между отзывами критиков и продажами XOne:', corr_criticXOne['critic_score'].corr(corr_criticXOne['total_sales']))

corr_userPC = actual_per_user_df[actual_per_user_df['platform'] == 'PC']
corr_criticPC = actual_per_critic_df[actual_per_critic_df['platform'] == 'PC']
print('Корреляция между отзывами пользователей и продажами PC:', corr_userPC['user_score'].corr(corr_userPC['total_sales']))
print('Корреляция между отзывами критиков и продажами PC:', corr_criticPC['critic_score'].corr(corr_criticPC['total_sales']))

corr_userPS4 = actual_per_user_df[actual_per_user_df['platform'] == 'PS4']
corr_criticPS4 = actual_per_critic_df[actual_per_critic_df['platform'] == 'PS4']
print('Корреляция между отзывами пользователей и продажами PS4:', corr_userPS4['user_score'].corr(corr_userPS4['total_sales']))
print('Корреляция между отзывами критиков и продажами PS4:', corr_criticPS4['critic_score'].corr(corr_criticPS4['total_sales']))

corr_userPSV = actual_per_user_df[actual_per_user_df['platform'] == 'PSV']
corr_criticPSV = actual_per_critic_df[actual_per_critic_df['platform'] == 'PSV']
print('Корреляция между отзывами пользователей и продажами PSV:', corr_userPSV['user_score'].corr(corr_userPSV['total_sales']))
print('Корреляция между отзывами критиков и продажами PSV:', corr_criticPSV['critic_score'].corr(corr_criticPSV['total_sales']))

corr_userWiiU = actual_per_user_df[actual_per_user_df['platform'] == 'WiiU']
corr_criticWiiU = actual_per_critic_df[actual_per_critic_df['platform'] == 'WiiU']
print('Корреляция между отзывами пользователей и продажами WiiU:', corr_userWiiU['user_score'].corr(corr_userWiiU['total_sales']))
print('Корреляция между отзывами критиков и продажами WiiU:', corr_criticWiiU['critic_score'].corr(corr_criticWiiU['total_sales']))
````

````python
plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_user3DS)
plt.title('Пользователи 3DS')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_critic3DS)
plt.title('Критики 3DS')

plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_userXOne)
plt.title('Пользователи XOne')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_criticXOne)
plt.title('Критики XOne')

plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_userPC)
plt.title('Пользователи PC')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_criticPC)
plt.title('Критики PC')

plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_userPS4)
plt.title('Пользователи PS4')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_criticPS4)
plt.title('Критики PS4')

plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_userPSV)
plt.title('Пользователи PSV')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_criticPSV)
plt.title('Критики PSV')

plt.figure(figsize=(5,5))
sns.scatterplot(x='user_score', y='total_sales', data = corr_userWiiU)
plt.title('Пользователи WiiU')
plt.figure(figsize=(5,5))
sns.scatterplot(x='critic_score', y='total_sales', data = corr_criticWiiU)
plt.title('Критики WiiU')
````

**Вывод:** как таковой четкой зависимости между отзывами критиков и суммой продаж нет, равно как между отзывами пользователей и суммой продаж. Однако корреляция между между отзывами критиков и суммой продаж выше. Можно сделать вывод, что к критикам прислушиваются больше, чем к пользователям. Ну и критиков больше слушают пользователи `Xbox One`.

## Проверка гипотез

````python
actual_per_df = (
    df
    .query('year_of_release > 2013')
)
````

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые:

```
H_0: Средние пользовательские рейтинги Xbox One = Средние пользовательские рейтинги PC`
H_a: Средние пользовательские рейтинги Xbox One ≠ Средние пользовательские рейтинги PC`
alpha = 0.05
```

<font color='DarkBlue'><b>Комментарий ревьюера</b></font><br>
<font color='DarkGreen'>👌 Хорошо, правильно сформулированы гипотезы.</font>

````python
from scipy import stats as st
````

````python
XOne = actual_per_df[(actual_per_df['platform'] == 'XOne') & (actual_per_df['user_score'] > 0)]['user_score']
PC = actual_per_df[(actual_per_df['platform'] == 'PC') & (actual_per_df['user_score'] > 0)]['user_score']

results = st.ttest_ind(XOne, PC)

alpha = 0.05 

print(results.pvalue) 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

На основании проверки p-уровня, вероятность, что между пользовательскими рейтингами Xbox One и PС есть связь, отсутствует.

### Средние пользовательские рейтинги жанров Action и Sports разные:

Любая гипотеза должна быть сформулирована так, чтобы использовать знак равенства. Например, нулевая гипотеза условно звучит так: "среднее рассматриваемой генеральной совокупности равно А". Перекладывая на случай с продажами игр гипотеза звучит так: _"среднее пользовательских рейтингов Action равен среднему пользовательских рейтингов Sports"._ Вероятность того, что гипотеза вероятна проверяется p-value при помощи метода ttest_ind.

```
H_0: Средние пользовательские рейтинги Action = Средние пользовательские рейтинги Sports`
H_a: Средние пользовательские рейтинги Action ≠ Средние пользовательские рейтинги Sports`
alpha = 0.05
```

````python
Action = actual_per_df[(actual_per_df['genre'] == 'Action') & (actual_per_df['user_score'] > 0)]['user_score']
Sport = actual_per_df[(actual_per_df['genre'] == 'Sports') & (actual_per_df['user_score'] > 0)]['user_score']

results2 = st.ttest_ind(Action, Sport)

alpha = 0.05 

print(results2.pvalue) 
if results2.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

Проверка p-уровня позволяет предположить, что между пользовательскими рейтингами жанра Action и Sports связи нет.
